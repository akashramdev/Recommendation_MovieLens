```python
import numpy as np
import pandas as pd
```

```python
movie = pd.read_csv("movie.csv")
```

```python
rating = pd.read_csv("rating.csv")
```

```python
imdb = pd.read_csv("imdb_top_1000.csv")
```

```python
imdb.head()
```

```python
movie['title'] = movie['title'].str.replace(r"\(\d+\)", "").str.strip()
```

```python
movie.head()
```

```python
movie.shape
```

```python
movies = movie.drop_duplicates(subset=['title'])
```

```python
movies.shape
```

```python
#movies.to_csv('movies.csv', index=False)
```

```python
imdb.head()
```

```python
imdb.shape
```

```python
imdb = imdb[['Series_Title', 'Overview']]
```

```python
movie_imdb = pd.merge(movies, imdb, left_on='title', right_on='Series_Title', how='left')
```

```python
movie_imdb.head()
```

```python
movie_imdb.isnull().sum()
```

```python
movie_imdb.shape
```

```python
df = movie_imdb.dropna(subset=['Series_Title'])
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.drop("Series_Title", axis = 1)
```

```python
df.shape
```

```python
rating.head()
```

```python
rating.shape
```

```python
final = df.merge(rating, on='movieId', how = 'left')
```

```python
final.head(10)
```

```python
final = final.sort_values(by='userId')
```

```python
final.head(10)
```

```python
final.shape
```

```python
final = final.drop_duplicates(subset=['movieId','userId'], keep = 'last')
```
